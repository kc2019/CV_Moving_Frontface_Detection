```python
import cv2
```

```python
face_cascade = cv2.CascadeClassifier('D:\KUNTAL\Python\python_notes\Image_Processing\haarcascade_frontalface_default.xml')
#eye_cascade = cv2.CascadeClassifier('D:\KUNTAL\Python\python_notes\Image_Processing\haarcascade_eye.xml')
```

```python
# Initiate video capture for video file
cap = cv2.VideoCapture(r'D:\KUNTAL\Python\python_notes\Image_Processing\video\people_face.mp4')
print(cap.isOpened())   # True = read video successfully. False - fail to read video.
```

```python
# loop runs if capturing has been initialized. 
while True: 
    # reads frames from a video 
    ret, frames = cap.read() 
    
    # convert to gray scale of each frames 
    gray = cv2.cvtColor(frames, cv2.COLOR_BGR2GRAY) 
    
    # Detects faces of different sizes in the input image 
    #cars = car_cascade.detectMultiScale(gray, 1.1, 1)
    faces = face_cascade.detectMultiScale(gray, 1.3, 5)
    #print("Found {0} faces!".format(len(faces)))
    
    # To draw a rectangle in each face 
    for (x,y,w,h) in faces: 
        #cv2.rectangle(frames,(x,y),(x+w,y+h),(0,0,255),2)
        cv2.rectangle(frames,(x,y),(x+w,y+h),(0,255,0),2) # RGB - green rectangle 
        
    # Display frames in a window  
    cv2.namedWindow("Moving People Face Detection", cv2.WND_PROP_FULLSCREEN)
    cv2.imshow('Moving People Face Detection', frames)
    
    # Wait for Esc key to stop 
    if cv2.waitKey(33) == 27: 
        break

# De-allocate any associated memory usage 
cv2.destroyAllWindows() 
```
